```python
import cv2
import mediapipe as mp
#mediapipe nos dara posiciones de estimacion
import numpy as np
#la primera es el dibujo de trazado mp,nos dara los trazados para dibujar las poses, 
#esto ahorita es igual a las soluciones de puntos de mp
mp_drawing=mp.solutions.drawing_utils
#aqui esta importando nuestro modelo de solucion de pose, el mismo mp tiene muchos modelos 
#de estimacion para manos cara etc. entonces aqui importamos el que esta relacionado a las poses
mp_pose=mp.solutions.pose
```

```python
#VIDEO FEED
#Primero capturamos nuestro medio de visualizacion, en este caso la camara frontal, esto se 
#configura con el numero puesto 0 que en este caso es la camara del webcam lo cual podria variar
cap=cv2.VideoCapture(0)

#aqui creamos una peque√±a instancia donde le pasamos el modelo colocando una confianza deteccion
#del 50% y un seguimiento del 50% en caso de necesitar ser mas estrictos con la deteccion puede
#aumentarse estas son las dos variables de entrada esto lo asignamos al nombre pose 
#para que podamos trabajar con esa linea pero con el nombre pose
with mp_pose.Pose(min_detection_confidence=0.5,min_tracking_confidence=0.5) as pose:
    #Se realiza un while para que la captura de la imagen de la camara sea constante
    while cap.isOpened():
        #y para leer los datos o almacenarlos se lo pone en una variable
        #donde ret es nuestra variable de retorno que no usaremos en este ejemplo
        #en este caso frame es aquella que capturara los datos de la imagen
        ret,frame=cap.read()
        
        #Recolor image RGB
        #aqui recoloreamos la imagen ya que cuando lo pasamos a mediapipe queremos que este 
        #en rgb, practicamente estamos reordenando las capas de color para pasarlas de manera correcta
        image=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        #decimos que no se escriba sobres esto ya que lo estamos coloreando
        image.flags.writeable=False
        
        #Aqui hacemos la deteccion
        results= pose.process(image)
        
        #Recoloreando a BGR, aqui recoloreamos en bgr y ponemos true para que se acepte dibujar sobre el 
        image.flags.writeable=True
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
        
       # print(results)
    
    
        (alto,ancho,canales)=image.shape
        img2=cv2.imread('media/treepose.jpg')
        img2=cv2.resize(img2,(ancho-200,(alto)))
        (altopp,anchopp,canalespp)=img2.shape
       # print(alto)
        prueba=np.concatenate((img2,image),axis=1)
        
        
        
        
        
        #Render detections
        #aqui se hace las detecciones se dibuja lo que estamos estimando sobre la imagen en la imagen
        #que aparece de la camara web, le pasamos tres primeros datos, nuestra imagen luego nuestros
        #resultados de la posicion de una persona
        ##las otras dos lineas de abajo, son especificaciones de dibujo la primera es el color que queremos para nuestros puntos
        # y la segunda es de nuestra conexion de los puntos igual especificamos el color
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                 mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))
                                
        #aqui mostramos la imagen cv2 nos da la manera de ponerlo en una ventana emergente
        #para visualizar nuestra imagen
        cv2.imshow('Mediapipe Feed', prueba)
        
        
        
        

        #En este caso estamos cerrando el loop de dos manera una en caso de que se aprete a tecla
        #para cerrar ventana de cv2 y como segunda en caso que se apriete la tecla q
        if cv2.waitKey(10) & 0xFF==ord('q'):
            break
    #aqui practicamente cerramos nuestra web cam y destruimos todas las ventanas emergentes rela
    #cionadas
    cap.release()
    cv2.destroyAllWindows()
```

```python
#aqui calculamos los angulos, lo que queremos es calcular el angulo entre nuestro hombro y nuestra mu√±eca
def calculate_angle(a,b,c):
    a=np.array(a)
    b=np.array(b)
    c=np.array(c)
    #no entiendo bien las matematicas aqui :V pero en gral se calculan los radianes 
    radians=np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    #y aqui se lo convierte en angulo
    angle=np.abs(radians*180.0/np.pi)
    #en caso de que el angulo sea mayor se lo resta ya que 180 es lo maximo a lo que va nuestro brazo 
    if angle >180.0:
        angle=360-angle
        
    return angle
```
